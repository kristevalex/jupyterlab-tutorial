```python
dt['Flactuation'] = dt.HighPrice - dt.LowPrice
dt['Diff'] = dt.ClosePrice - dt.OpenPrice
dt['PrevDiff'] = dt.Diff.shift(1)
dt['PrevFlactuation'] = dt.Flactuation.shift(1)
```

```python
x = dt.PrevDiff
y = dt.Diff
x1 = dt.PrevFlactuation
y1 = dt.Diff
plt.figure(figsize=(12, 8), dpi=80)
plt.grid(True, which='both')
plt.scatter(x, y, alpha=0.1, color = 'blue')
plt.scatter(x1, y1, alpha=0.1, color = 'red')
```
